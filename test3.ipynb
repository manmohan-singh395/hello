```python
# factorial
def fact(n):
    result = 1
    for i in range(1,n+1):
        result *= i
    return result
fact(5)

```

```python
def filter_dynamic(data, threshold):
    result = []
    for num in data:
        if num > threshold and num % 2 == 0:
            result.append(num)
    return result

print(filter_dynamic([2, 5, 8, 11, 12, 3, 6], 5)) 
```

```python
store1 = {"apple": 10, "banana": 5, "orange": 8}
store2 = {"banana": 10, "apple": 2, "grapes": 5}

merged = {}
for item in store1:
    merged[item] = store1[item]
for item in store2:
    if item in merged:
        merged[item] = merged[item] + store2[item] 
    else:
        merged[item] = store2[item] 

print(merged)
```

```python
def flexible_multiply(*args):
    if not args: 
        return 0
    product = 1
    for num in args:
        product *= num
    return product

print(flexible_multiply(2, 3, 4)) 
print(flexible_multiply()) 
```

```python
def safe_read(file_name):
    try:
        with open(file_name, 'r') as f:
            print(f.read())
    except FileNotFoundError:
        with open(file_name, 'w') as f:
            f.write("Default Content")
        print("File not found. Created with default content.")

safe_read("data.txt")
```

```python
def is_palindrome(text):
    text = text.replace("", "").lower()
    return text == text[::-1]

print(is_palindrome("Racecar"))
```

```python
def create_file(name):
    with open("database.txt","a") as f:
        f.write(name + "\n")
        print("file created successfully")
create_file("aman")
create_file("manmohan")
create_file("akshay")
create_file("himanshi")

```

```python
try:
    with open("test.txt" , "r") as f:
        content = f.read()
        print(content)
except FileNotFoundError:
    print("File not found")
```

```python
try:
    with open("test.txt" , "x") as f:
        f.write("hello world!")
        print("active mode on")

except FileNotFoundError:
    print("File not found")
```

```python
#,accepted value
import csv

data = [["Name","Course"],["Aman","pyton"],["Man Mohan","jawa"],["Akhay","node"],["Hianshi","react"]]

with open("data.csv","w") as f:
    writer = csv.writer(f)
    writer.writerows(data)

with open("data.csv","r") as f:
    reader = csv.reader(f)
    for row in reader:
        print(row)

```

```python
#error handling
try:
    num = int(input("enter a number :"))
    result = 10/num
except ValueError:
    print("pease enter a valid number")
except ZeroDivisionError:
    print("you can't divide by zero")
else:
    print(f"Result : {result}")
finally:
    print("programme ended")
```
